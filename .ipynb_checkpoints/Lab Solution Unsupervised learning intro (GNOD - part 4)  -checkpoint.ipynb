```python
#initialize empty lists
titles = []
artists = []


# define the number of iterations of our for loop
num_iter = len(soup.select("div.feed-item"))

titles_list = soup.select("cite.title") #titles
artists_list = soup.select("em.artist") # artists
# iterate through the result set and retrive all the data
for i in range(num_iter):
    titles.append(titles_list[i].get_text())
    artists.append(artists_list[i].get_text())

print(titles)
print(artists)
```

```python
song_title = []
artist = []
page = len(soup.select("div.chart-content.col-xs-12.col-sm-8"))

titles_list = soup.select("div.chart-content.col-xs-12.col-sm-8 > p > cite") #titles
artists_list = soup.select("div.chart-content.col-xs-12.col-sm-8 > p > em") #artist

for i in range(page):
    song_title.append(titles_list[i].get_text())
    artist.append(artists_list[i].get_text())


print(song_title)
artist
    

```

## Music recommender

```python
import random

def songrecommender():
    artist = input("Enter who sang it: ")
    song = input("Enter name of song: ")
    
    for i in range(len(top100.values)):
        if (artist in top100.values[i][0]) & (song in top100.values[i][1]):
            recommendation = top100[(top100['artist'] != artist) & (top100['song_title'] != song)].iloc[random.randint(0,99)]
            return "We recommend: " +recommendation.artist+"'s" " " + recommendation.song_title 
        
    for i in range(len(top100.values)): 
        if (artist not in top100.values[i][0]) & (song not in top100.values[i][1]):
            return "Sorry, no recommendations available"
```

```python
secrets_file = open("secrets.txt","r")
string = secrets_file.read()
string.split('\n')
secrets_dict={}
for line in string.split('\n'):
    if len(line) > 0:
        secrets_dict[line.split(':')[0]]=line.split(':')[1].strip()

sp = spotipy.Spotify(auth_manager=SpotifyClientCredentials(client_id=secrets_dict['cid'],
                                                           client_secret=secrets_dict['csecret']))
```

# Scaling numerical features

```python

from sklearn.preprocessing import StandardScaler
#X_prep = StandardScaler().fit_transform(numerical_spotify_songs)
#pd.DataFrame(X_prep,columns=numerical_spotify_songs.columns).head()

transformer = StandardScaler().fit(numerical_spotify_songs)
X_prep = transformer.transform(numerical_spotify_songs)
noralised_numericals = pd.DataFrame(X_prep ,columns=numerical_spotify_songs.columns)
noralised_numericals
```

##### finding the optimum K

```python
import numpy as np
from sklearn.cluster import KMeans

K = range(2, 20)
inertia = []

for k in K:
    kmeans = KMeans(n_clusters=k,
                    random_state=1234)
    kmeans.fit(X_prep)
    inertia.append(kmeans.inertia_)

import matplotlib.pyplot as plt
%matplotlib inline

plt.figure(figsize=(16,8))
plt.plot(K, inertia, 'bx-')
plt.xlabel('k')
plt.ylabel('inertia')
plt.xticks(np.arange(min(K), max(K)+1, 1.0))
plt.title('Elbow Method showing the optimal k')
```

```python
from sklearn.metrics import silhouette_score
K = range(2, 20)
silhouette = []

for k in K:
    kmeans = KMeans(n_clusters=k,
                    random_state=1234)
    kmeans.fit(X_prep)
    silhouette.append(silhouette_score(X_prep, kmeans.predict(X_prep)))


plt.figure(figsize=(16,8))
plt.plot(K, silhouette, 'bx-')
plt.xlabel('k')
plt.ylabel('silhouette score')
plt.xticks(np.arange(min(K), max(K)+1, 1.0))
plt.title('Slhouette score showing the optimal k')
```

# testing for k

```python
spotify_songs['clusters'] = pred
spotify_songs
```

# Now GNOD song recommender

```python
import random

def songrecommender():
    artist = input("Enter who sang it: ")
    song = input("Enter name of song: ")
    
    for i in range(len(top100.values)):
        if (artist in top100.values[i][0]) & (song in top100.values[i][1]):
            recommendation = top100[(top100['artist'] != artist) & (top100['song_title'] != song)].iloc[random.randint(0,99)]
            return "We recommend: " +recommendation.artist+"'s" " " + recommendation.song_title 
        
    #songs not in top100 playlist
    for i in range(len(spotify_songs.values)): 
        if (song not in top100.values[i][0]) & (artist not in top100.values[i][1]):
                #get features from spotify
                track_id = sp.search(q='artist:' + artist + ' track:' + song, type='track') #found on stackov
                songuri = track_id["tracks"]["items"][1]["uri"] 
                get_feat = sp.audio_features(songuri)
                
                #drop categoricals
                featdf = pd.DataFrame(get_feat)
                featdf = featdf.drop(['type','id','uri','track_href','analysis_url'], axis=1)
                
                #scale features
                featnorm = transformer.transform(featdf)
                featnorm = pd.DataFrame(featnorm,columns=featdf.columns)
                
                #get cluster and recommend song in cluster
                clustnum = kmeans.predict(featnorm)[0]
                cluster = spotify_songs[spotify_songs['clusters'] == clustnum].reset_index()
                recommendation = cluster[(cluster['song'] != song) & (cluster['artist'] != artist)].iloc[random.randint(0,len(cluster))]
                return "We recommend: " + recommendation.song + " by " + recommendation.artist + ".  " "Link to song: " + "https://open.spotify.com/track/" + recommendation.id + " "
```
